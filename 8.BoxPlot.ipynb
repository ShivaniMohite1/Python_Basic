# Box Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x)
plt.show()
```

#### make it horizontal

```python
x = [10,20,30,40,50,60,70]
plt.subplot(2,2,1)
plt.boxplot(x,vert= False)


#change width of boxplot

plt.subplot(2,2,2)
plt.boxplot(x, vert=False, widths=0.5)


#to fill the box with color
plt.subplot(2,2,3)
plt.boxplot(x, vert=True, patch_artist= True)


# Show Mean value : 
plt.subplot(2,2,4)
plt.boxplot(x,showmeans = True)


plt.show()
```

### Outliers

```python
#Outlier
x = [10,20,30,40,50,60,70,120]
plt.subplot(1,3,1)
plt.boxplot(x,showmeans=1)



plt.subplot(1,3,2)
x1 =[10,20,30,40,50,60,70,120,130] 
plt.boxplot(x1,showmeans=1)



#change color of outlier 
x =[10,20,30,40,50,60,70,120]
plt.subplot(1,3,3)
plt.boxplot(x,showmeans=1,sym="k d")


plt.tight_layout()
plt.show()
```

```python
#color Formatting of boxplot , change box color and cap colors:
x = [10,20,30,40,50,60,70]
plt.subplot(1,4,1)
plt.boxplot(x, boxprops = dict(color = "r"))

#change cap colors using capprops 
plt.subplot(1,4,2)
plt.boxplot(x, boxprops = dict(color = "g"), capprops= dict(color = 'Orange'), showmeans=1)


plt.subplot(1,4,3)
plt.boxplot(x, boxprops = dict(color = "b"), capprops= dict(color = 'b'),whiskerprops=dict(color = 'g'), )


plt.tight_layout()
plt.show()
```

```python
x = [10,20,30,40,50,60,70]
y = [21,1,25,46,37,28,69,70,36,55,120]

z = (x,y)





plt.boxplot(z, labels=["box1",'Box2'], showmeans=1)
plt.show()
```

Boxplot by random number generation

```python
import numpy as np

# Generate random data
np.random.seed(42)
data1 = np.random.randint(0, 100, 100)
data2 = np.random.randint(2, 130, 100)
data3 = np.random.normal(-2, 50, 100)

# Combine the data
data = [data1, data2, data3]
```

```python
import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots()

# Create a boxplot
ax.boxplot(data, labels=['Category 1', 'Category 2', 'Category 3'])

# Add a title and labels
ax.set_title('Boxplot Example')
ax.set_xlabel('Categories')
ax.set_ylabel('Values')

# Show the plot
plt.show()
```
