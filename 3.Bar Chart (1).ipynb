```python
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children)
```

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color=colors)
```

```python
# horizontal bar chart
plt.bar(colors,children, color = colors)

plt.title("title")
plt.xlabel('colors')
plt.ylabel('no.of childrens')

plt.show()
```

```python
# color and label
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0]) -0.2,df['2015'],width=0.2,color='b')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='darkorange')
plt.bar(np.arange(df.shape[0]) +0.2,df['2017'],width=0.2,color='g')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```
