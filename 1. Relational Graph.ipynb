-Bivariate Analysis

-Plots under this section

1)scatterplot

2)lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips["total_bill"],tips["tip"])  # matplotlib 
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
# scatter plot -> axes level function
sns.scatterplot(data=tips, x='total_bill', y='tip',hue ='time',style='sex',size='size')
```

```python
# relplot -> figure level -> square shape
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter',hue='sex',style='time',size='size')
plt.show()
```

```python
# line plot
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# axes level function
sns.lineplot(data=temp_df, x='year', y='lifeExp')
```

```python
# using relpplot
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country', style='continent')
```

```python
sns.lineplot(data=temp_df, x='year', y='lifeExp', hue='country')   #axis level function 
```

### Facetplot

```python
# facet plot -> figure level function -> work with sns.relplot
# it will not work with sns.scatterplot and sns.lineplot
sns.relplot(kind='scatter',data=tips, x='total_bill', y='tip', col='day', row='sex')
```

```python
# col wrap
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year')
```

```python
# col wrap 
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap = 5)
```
