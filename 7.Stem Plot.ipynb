# Stem Plot

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,6)
x
```

```python
np.random.seed(1)
y = np.random.randint(1,20,5)
y
```

```python
plt.stem(x,y)
plt.title("stemplot")
```

```python
plt.subplot(2,1,1)
plt.stem(x,y)
plt.title("stemplot")

# making line dotted we use linefmt function
plt.subplot(2,1,2)
plt.stem(x,y,linefmt= ':')
plt.title("dotted line")


plt.tight_layout()
```

```python
# we can edit marker also using markerfmt
plt.subplot(2,1,1)
plt.stem(x,y,linefmt= '--',markerfmt="r p")
plt.title("change marker")


#change base line color using basefmt
plt.subplot(2,1,2)
plt.stem(x,y, basefmt="g")
plt.title("stemplot")

plt.tight_layout()
```

```python

#CHange orientation
plt.subplot(2,1,1)
plt.stem(x,y, basefmt="g",orientation = "horizontal")
plt.title("stemplot")

plt.tight_layout()
```
