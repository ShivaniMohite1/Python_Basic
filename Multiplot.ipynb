```python
#import matplotlib.pyplot as plt
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.catplot(data= tips, x="sex", y="total_bill", kind = "violin", row= "time", col= "day") 
plt.show()
```

#by using facetgrid we can create more flexible facet column charts like above 

```python
g = sns.FacetGrid(data= tips, row= "time", col= "day")
plt.show()
```

```python
g = sns.FacetGrid(data= tips, row= "time", col= "day", hue = "sex")
g.map(sns.violinplot, "sex", "total_bill")
g.add_legend()
plt.show()
```

# Pairplot

gives pairwise relation 

```python
d = sns.load_dataset("iris")
d
```

```python
sns.pairplot(d)
plt.show()
```

```python
sns.pairplot(d, hue = "species")
plt.show()
```

# PairGrid

- Customisable Pairplot

```python
P = sns.PairGrid(data = d)
P.map(sns.scatterplot)
plt.show()
```

```python
P = sns.PairGrid(data = d, hue = "species")
P.map_diag(sns.histplot)
P.map_offdiag(sns.lineplot)
plt.show()
```

```python
#map_diag and map_offdiag
P = sns.PairGrid(data = d, hue = 'species')
P.map_diag(sns.boxplot)
P.map_offdiag(sns.histplot)
plt.show()
```

```python
#map_upper  and map_lower
P = sns.PairGrid(data = d, hue = 'species')
P.map_diag(sns.histplot)
P.map_upper(sns.scatterplot)
P.map_lower(sns.lineplot)
plt.show()
```

#vars 
we give vars parameter to plot the graph between user given columns 

```python
P = sns.PairGrid(data = d, hue = 'species', vars=['petal_length','petal_width'])
P.map_diag(sns.histplot)
P.map_upper(sns.scatterplot)
P.map_lower(sns.lineplot)
plt.show()
```

# Jointplot

```python
sns.jointplot(data = tips, x='tip', y ='total_bill')
plt.show()
```

```python
sns.jointplot(data = tips, x='tip', y ='total_bill', kind = 'reg')
plt.show()
```

```python
sns.jointplot(data = tips, x='tip', y ='total_bill', kind = 'hex')
plt.show()
```

```python
sns.jointplot(data = tips, x='tip', y ='total_bill', kind = 'hist' )
plt.show()
```

```python
g = sns.JointGrid(data = tips, x='tip', y ='total_bill')
g.plot(sns.regplot, sns.kdeplot)
plt.show()
```
