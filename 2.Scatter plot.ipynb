```python
df = pd.read_csv("batter.csv")
df
```

```python
plt.scatter(df['avg'],df['strike_rate'], color='green', marker='P')
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

```python
# size
tips = sns.load_dataset('tips')

tips
```

```python
# slower
plt.scatter(tips['total_bill'],tips['tip'],s =tips['size']*20)
```

```python
# scatterplot using plt.plot
# faster
plt.plot(tips['total_bill'],tips['tip'],'s')
```
