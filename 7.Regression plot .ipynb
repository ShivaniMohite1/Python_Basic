#### What is Regression? 

Regression, in simple words, is a statistical method used to understand and model the relationship between two or more variables. It helps us predict one variable (the "dependent" or "target" variable) based on the values of one or more other variables (the "independent" or "predictor" variables).




magine you want to figure out how one thing affects another. For instance, you might wonder how the number of hours people study (the predictor variable) influences their exam scores (the target variable). You collect data from a group of students, recording both the study hours and their corresponding exam scores. To understand this relationship, you can use regression.

In regression, you look for a pattern or a mathematical equation that best describes how changes in study hours are associated with changes in exam scores. This equation might say something like, "For every additional hour of study, the exam score is expected to increase by X points."

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
tips = sns.load_dataset("tips")
tips
```

```python

sns.scatterplot(data = tips, x="total_bill", y = "tip")
plt.show()
```

```python
# regplot => axes level function
sns.regplot(data = tips, x="total_bill", y = "tip")
plt.show()

#Light bluish shade near to line shows error in line so we can also say that upto that range line may be deviated
```

```python
# lmplot ==> figure level function
sns.lmplot(data = tips, x="total_bill", y = "tip")
plt.show()
```

```python
sns.lmplot(data = tips, x="total_bill", y = "tip", hue = 'sex')  # hue parameter not working on axes level function regplot
plt.show()
```
