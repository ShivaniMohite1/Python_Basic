### Heatmap 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
import pandas as pd

data = {
    'Hours Studied': [2, 3, 1, 4, 5, 2, 3, 4, 5, 1],
    'Test Score': [65, 70, 50, 80, 85, 60, 75, 85, 90, 55]
}

df = pd.DataFrame(data)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

This code will generate a heatmap that visually represents the correlation between the "Hours Studied" and "Test Score" variables in your dataset. You can explain to your students that:

The color intensity in the cells represents the strength and direction of the correlation.
Positive correlations are indicated by warmer colors (e.g., red), while negative correlations are indicated by cooler colors (e.g., blue).
The values inside the cells (annot=True) show the actual correlation coefficients.
In this dataset, you can expect a positive correlation between the number of hours studied and the test scores, which should be reflected in the heatmap

```python
gap = px.data.gapminder()   
gap
```

```python
gap1 = gap.pivot(index ="country", columns="year", values="lifeExp")
gap1
```

```python
#heatmap => Axis level function

sns.heatmap(gap1)
plt.show()
```

```python
# As it is axis level function we an adjust figure size as same as matplotlib
plt.figure( figsize = (20,15))
sns.heatmap(gap1)
plt.show()
```

```python
gap2 = gap[gap['continent']== "Europe"].pivot(index ="country", columns="year", values="lifeExp" )
gap2
```

```python
plt.figure( figsize = (20,20))
sns.heatmap(gap2, annot=True)  # annote used to show the data lables
plt.show()
```

```python
plt.figure( figsize = (20,20))
sns.heatmap(gap2, annot=True, linewidth = 0, cmap="viridis")  # linewidth used to increase gap between two heatmaps to improve  readability
plt.show()
```
